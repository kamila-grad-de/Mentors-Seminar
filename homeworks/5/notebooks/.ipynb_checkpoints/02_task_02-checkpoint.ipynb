```python
import pandas as pd
from sqlalchemy import create_engine
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
POSTGRES_HOST = "postgres"
POSTGRES_PORT = 5432
POSTGRES_DB = "analytics"
POSTGRES_USER = "user"
POSTGRES_PASSWORD = "pass"

pg_url = (
    f"postgresql+psycopg2://{POSTGRES_USER}:{POSTGRES_PASSWORD}"
    f"@{POSTGRES_HOST}:{POSTGRES_PORT}/{POSTGRES_DB}"
)

engine = create_engine(pg_url)
```

```python
targets = pd.read_sql("SELECT * FROM well_targets", engine)
telemetry = pd.read_sql("SELECT * FROM well_telemetry", engine)

targets["date"] = pd.to_datetime(targets["date"]).dt.date
telemetry["timestamp"] = pd.to_datetime(telemetry["timestamp"])
telemetry["date"] = telemetry["timestamp"].dt.date
```

Агрегация telemetry по дням (всего 2 строки остается, я не понимаю что я делаю не так)

```python
telemetry_agg = (
    telemetry
    .groupby(["well_id", "date"], as_index=False)
    .agg(
        pressure=("pressure_out", "mean"),
        temperature=("temperature", "mean"),
        power=("pump_speed_rpm", "mean"),
        pump_work_time=("record_id", "count"),
    )
)

df = targets.merge(telemetry_agg, on=["well_id", "date"], how="inner")
df.head()
```

Обучить модель: Linear Regression / RandomFores

```python
features = [
    "pressure",
    "temperature",
    "power",
    "pump_work_time",
]

target = "daily_oil_ton"

X = df[features]
y = df[target]

lr = LinearRegression()
lr.fit(X, y)

rf = RandomForestRegressor()
rf.fit(X, y)
```

Оценить: MAE / RMSE

```python
lr_preds = lr.predict(X);
rf_preds = rf.predict(X);

def print_metrics(preds, name):        
    print(name)
    print("MAE:", mean_absolute_error(y, preds))
    print("RMSE:", mean_squared_error(y, preds) ** 0.5)

print_metrics(lr_preds, "LinearRegression")
print_metrics(rf_preds, "RandomForestRegressor")
```

Сделать прогноз

```python
df["lr_preds"] = lr.predict(X);
df["rf_preds"] = rf.predict(X);

df
```
