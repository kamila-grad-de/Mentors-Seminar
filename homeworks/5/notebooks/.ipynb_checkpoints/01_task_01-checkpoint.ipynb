```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
POSTGRES_HOST = "postgres"
POSTGRES_PORT = 5432
POSTGRES_DB = "analytics"
POSTGRES_USER = "user"
POSTGRES_PASSWORD = "pass"

pg_url = (
    f"postgresql+psycopg2://{POSTGRES_USER}:{POSTGRES_PASSWORD}"
    f"@{POSTGRES_HOST}:{POSTGRES_PORT}/{POSTGRES_DB}"
)

engine = create_engine(pg_url)
```

```python
wells = pd.read_sql("SELECT * FROM wells", engine)
production = pd.read_sql("SELECT * FROM production", engine)

production["date"] = pd.to_datetime(production["date"])
```

Общая добыча по дням

```python
daily_production = (
    production
    .groupby("date", as_index=False)
    .agg(
        total_oil_ton=("oil_ton", "sum"),
        total_gas_m3=("gas_m3", "sum"),
        total_energy_kwh=("energy_kwh", "sum"),
    )
    .sort_values("date")
)

daily_production.head()
```

KPI по скважинам (средний дебит и % простоя)

```python
well_kpi = (
    production
    .groupby("well_id", as_index=False)
    .agg(
        avg_oil_ton=("oil_ton", "mean"),
        avg_temperature=("temperature", "mean"),
        avg_pressure=("pressure", "mean"),
        total_downtime=("downtime_hours", "sum"),
    )
)

well_kpi["downtime_percent"] = well_kpi["total_downtime"] / (30 * 24) * 100
well_kpi = well_kpi.merge(wells, on="well_id", how="left")

well_kpi[["avg_oil_ton", "downtime_percent"]].head()
```

Лучшие и худшие скважины

```python
sorted_wells = well_kpi.sort_values("avg_oil_ton", ascending=False)
sorted_wells[["well_id", "name", "avg_oil_ton"]].head()
```
