```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
POSTGRES_HOST = "postgres"
POSTGRES_PORT = 5432
POSTGRES_DB = "analytics"
POSTGRES_USER = "user"
POSTGRES_PASSWORD = "pass"

pg_url = (
    f"postgresql+psycopg2://{POSTGRES_USER}:{POSTGRES_PASSWORD}"
    f"@{POSTGRES_HOST}:{POSTGRES_PORT}/{POSTGRES_DB}"
)

engine = create_engine(pg_url)
```

```python
deliveries = pd.read_sql("SELECT * FROM deliveries", engine)
```

Определить факторы задержек:
- погода
- расстояние
- водитель

```python
weather_delay = (
    deliveries
    .groupby("weather_conditions", as_index=False)
    .agg(avg_delay_hours=("delay_hours", "mean"))
    .sort_values("avg_delay_hours", ascending=False)
)

weather_delay
```

```python
deliveries[["distance_km", "delay_hours"]].corr()
```

```python
driver_delay = (
    deliveries
    .groupby("driver_id", as_index=False)
    .agg(avg_delay_hours=("delay_hours", "mean"))
    .sort_values("avg_delay_hours", ascending=False)
)

driver_delay
```

Проанализировать стоимость: cost / km

```python
deliveries["cost_per_km"] = deliveries["cost_usd"] / deliveries["distance_km"]

route_cost = (
    deliveries
    .groupby(["source", "destination"], as_index=False)
    .agg(avg_cost_per_km=("cost_per_km", "mean"))
    .sort_values("avg_cost_per_km", ascending=False)
)

route_cost.head()
```

Оптимизировать маршруты

```python
optimized_routes = (
    deliveries
    .groupby(["source", "destination"], as_index=False)
    .agg(
        avg_delay=("delay_hours", "mean"),
        avg_cost=("cost_usd", "mean"),
        avg_cost_per_km=("cost_per_km", "mean"),
    )
    .sort_values(["avg_delay", "avg_cost_per_km"])
)

optimized_routes.head()
```
