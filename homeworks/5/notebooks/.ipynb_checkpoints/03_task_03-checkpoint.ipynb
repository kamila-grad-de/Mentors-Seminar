```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
from scipy.stats import zscore
from sklearn.ensemble import IsolationForest, RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
POSTGRES_HOST = "postgres"
POSTGRES_PORT = 5432
POSTGRES_DB = "analytics"
POSTGRES_USER = "user"
POSTGRES_PASSWORD = "pass"

pg_url = (
    f"postgresql+psycopg2://{POSTGRES_USER}:{POSTGRES_PASSWORD}"
    f"@{POSTGRES_HOST}:{POSTGRES_PORT}/{POSTGRES_DB}"
)

engine = create_engine(pg_url)
```

```python
pump_sensors = pd.read_sql("SELECT * FROM pump_sensors", engine)
pump_failures = pd.read_sql("SELECT * FROM pump_failures", engine)

pump_sensors["timestamp"] = pd.to_datetime(pump_sensors["timestamp"])
pump_failures["failure_date"] = pd.to_datetime(pump_failures["failure_date"])
```

Найти аномалии: z-score / isolation forest

```python
features = [
    "temperature",
    "vibration",
    "current",
    "rpm",
]

X = pump_sensors[features]

z_scores = np.abs(zscore(X))
if_scores = IsolationForest().fit_predict(X)

pump_sensors["zscore_anomaly"] = z_scores.apply(lambda x: any(x > 3), axis=1)
pump_sensors["if_anomaly"] = if_scores == -1

pump_sensors[["zscore_anomaly", "if_anomaly"]].head()
```

Определить: признаки перед отказом

```python
before_failure = []
pump_sensors["failure_target"] = 0

for _, failure in pump_failures.iterrows():
    pump_id = failure["pump_id"]
    
    failure_date = failure["failure_date"]
    
    start_time = failure_date - pd.Timedelta(hours=24)
    mask = ((pump_sensors["pump_id"] == pump_id) & (pump_sensors["timestamp"] >= start_time) & (pump_sensors["timestamp"] <= failure_date))
    subset = pump_sensors[mask]
    
    before_failure.append(subset)
    
    pump_sensors.loc[mask, "failure_target"] = 1

before_failure_df = pd.concat(before_failure, ignore_index=True)
before_failure_df[features].mean()
```

Построить модель: вероятность отказа

```python
X = pump_sensors[features]
y = pump_sensors["failure_target"]

rf = RandomForestClassifier()
rf.fit(X, y)

pump_sensors["failure_proba"] = rf.predict_proba(X)[:, 1]
pump_sensors[["record_id", "failure_target", "failure_proba"]].tail(10)
```
